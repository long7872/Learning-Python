Bài 17: Sử dụng cấu trúc từ điển, viết một chương trình nhập một câu, in số chữ cái và chữ số trong câu đó. Giả sử đầu vào chuỗi: hello world! 123

```python
class SumAlnum:
    def __init__(self):
        self.dc = {}
        self.dc['alpha'] = 0
        self.dc['number'] = 0
    def insertStrToDict(self, st: str):
        for c in st:
            if c.isnumeric(): self.dc['number'] += 1
            elif c.isalpha(): self.dc['alpha'] += 1
    def printAlnumCount(self):
        print("Alpha:", self.dc['alpha'], "\nNumber:", self.dc['number'])
        

st = input("Input string: ")
sumAlnum = SumAlnum()
sumAlnum.insertStrToDict(st)
sumAlnum.printAlnumCount()
```

Bài 18: Sử dụng cấu trúc từ điển, viết một chương trình nhập một câu, in số chữ hoa và số chữ thường trong câu đó. Giả sử đầu vào chuỗi: Hello World

```python
class SumUpLower:
    def __init__(self):
        self.dc = {}
        self.dc['upper'] = 0
        self.dc['lower'] = 0
    def insertStrToDict(self, st: str):
        for c in st:
            if c.isupper(): self.dc['upper'] += 1
            elif c.islower(): self.dc['lower'] += 1
    def printAlnumCount(self):
        print("Upper:", self.dc['upper'], "\nLower:", self.dc['lower'])
        

st = input("Input string: ")
sumUpLower = SumUpLower()
sumUpLower.insertStrToDict(st)
sumUpLower.printAlnumCount()
```

Bài 19: Hiển thị 10 từ có tuần suất xuất hiện nhiều nhất (phổ biến nhất)

```python
class FreqWord:
    def __init__(self):
        self.dc = {}
    def calFreqFromFile(self, file_path):
        file = open(file_path, 'r')
        for line in file:
            words = line.split()
            for word in words:
                self.dc[word] = self.dc.get(word, 0) + 1
    def sortDictAndPrint(self):
        ls = []
        for tup in self.dc.items():
            ls.append(tup)
        ls.sort(key=lambda item: item[1], reverse=True)
        
        self.printList(ls)
    def printList(self, ls: list):
        for item in ls[:10]:
            print(item[0], ":", item[1])
        

freq = FreqWord()
freq.calFreqFromFile("source/chap_2/file_input.txt")
freq.sortDictAndPrint()
```

Bài 20: Viết một chương trình tính và in kết quả giai thừa của một số nhập vào từ bàn phím.     

```python
class Fac:
    def calculate(self, n):
        self.n = n
        self.fac = 1
        for i in range(2, n+1):
            self.fac *= i
    def printFactorial(self):
        print("Factorial of", self.n, "is:", self.fac)
        
fac = Fac()
fac.calculate(4)
fac.printFactorial()
```

Bài 21: Định nghĩa một hàm có thể tạo và in list chứa các giá trị bình phương của các số từ 1 đến 20 (tính cả 1 và 20). Sau đó thực thi hàm.

Gợi ý:
- Sử dụng toán tử ** để lấy giá trị bình phương.
- Sử dụng range() cho vòng lặp.
- Sử dụng phương thức append() để thêm giá trị vào danh sách.

```python
class CalSquare:
    ls = []
    @classmethod
    def calculateSquare(cls, lower_bound = 1, upper_bound = 20):
        cls.ls = []
        for i in range(lower_bound, upper_bound+1):
            cls.ls.append(str(i**2))
    @classmethod
    def printList(cls):
        print(', '.join(cls.ls))

CalSquare.calculateSquare()
CalSquare.printList()
```

Bài 22: Định nghĩa một hàm có thể tạo danh sách chứa các giá trị bình phương của các số từ 1 đến 20 (bao gồm cả 1 và 20) và in 5 phần tử đầu tiên cùng với 5 phần tử cuối cùng trong danh sách.

Gợi ý:
- Sử dụng toán tử ** để lấy giá trị bình phương.
- Sử dụng range() cho vòng lặp.
- Sử dụng phương thức append() để thêm giá trị vào danh sách.
- Sử dụng [n1:n2] để trích danh sách và in

```python
class CalSquare:
    ls = []
    @classmethod
    def calculateSquare(cls, lower_bound = 1, upper_bound = 20):
        cls.ls = []
        for i in range(lower_bound, upper_bound+1):
            cls.ls.append(str(i**2))
    @classmethod
    def printList5Begin5Last(cls):
        ls = cls.ls[:5] + cls.ls[-5:]
        print(', '.join(ls))

CalSquare.calculateSquare()
CalSquare.printList5Begin5Last()
```

Bài 23: Định nghĩa một hàm Check_chan(n) kiểm tra một số có phải là số chẵn hay không. Yêu cầu:
- Hàm được luu vào tệp Kiem_tra.py
- Tạo tệp Test.py để sử dụng hàm này nhằm kiểm tra một số nhập vào từ bàn phím có phải là số chẳn hay không.

```python
class EvenOdd:
    def __init__(self, n):
        self.n = n
    @staticmethod
    def checkEven(n):
        if n % 2 == 0: 
            return True
        else: return False

n = int(input())
if EvenOdd.checkEven(n): print(n, "is Even")
else: print(n, "is odd")
```

Bài 25: Tạo tệp tu_dien.py để định nghĩa:
- 01 hàm Tao_TD(Max) để tạo 1 từ điển, chứa các key là số từ 1 đến Max (số Max nhập vào từ bàn phím) và các giá trị là bình phương của key. 
- 01 Hàm Print_Item(TD) để in các phần tử của từ điển.
- 01 Hàm Print_Key(TD) để in các giá trị của các phần tử từ điển
- 01 Hàm Print_Value(TD) để in giá trị của các phần tử từ điển 
- Tạo tệp Test_TD.py để sử dụng các hàm này.

```python
class ManDict:
    def __init__(self):
        self.dc = {}
    def create_dict(self, max):
        for i in range(1, max+1):
            self.dc[i] = i**2
    def print_item(self):
        for tup in self.dc.items():
            print(tup, end=" ")
        print()
    def print_key(self):
        print(', '.join(map(str, self.dc.keys())))
    def print_value(self):
        print(', '.join(map(str, self.dc.values())))

md = ManDict()
md.create_dict(10)
md.print_item()
md.print_key()
md.print_value()
```
