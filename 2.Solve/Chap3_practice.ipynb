Example Polymorphism 3:

```python
from math import pi

class Shape:
    def __init__(self, name):
        self.name = name
    def area(self):
        pass
    def fact(self):
        return "A two-dimensional shape."
    def __str__(self):
        return self.name

class Square(Shape):
    def __init__(self, length):
        super().__init__("Square")
        self.length = length
    def area(self):
        return self.length**2
    def fact(self):
        return "Square have each angle equal to 90 degrees."

class Circle(Shape):
    def __init__(self, radius):
        super().__init__("Circle")
        self.radius = radius
    def area(self):
        return pi*self.radius**2


a = Square(4)
b = Circle(7)
print(b)
print(b.fact())
print(a.fact())
print(b.area())
```

Example Abstraction Interface:

/49

```python
from abc import ABC, abstractmethod
from math import pi

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
    @abstractmethod
    def perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius) -> None:
        super().__init__()
        self.radius = radius
    def area(self):
        return pi * self.radius ** 2
    def perimeter(self):
        return 2 * pi * self.radius

class Rectangle(Shape):
    def __init__(self, width, height) -> None:
        super().__init__()
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def perimeter(self):
        return (self.width + self.height) * 2


a = Circle(4)
b = Rectangle(7, 8)
print(b)
print(a.area())
print(a.perimeter())
print(b.area())
print(b.perimeter())
```

Example Encapsulation:

/50

```python
class Student:
    def __init__(self, univer):
        self.__univer = univer
    
    @property
    def univer(self):
        return self.__univer

a = Student("VKU")
a.__univer = "VKU University"
print(a.univer)
```

```python
class Student:
    def __init__(self, univer):
        self.__univer = univer
    
    @property
    def univer(self):
        return self.__univer
    
    @univer.setter
    def univer(self, univer):
        self.__univer = univer

a = Student("VKU")
a.univer = "VKU University"
# print(a.__univer) # 'Student' object has no attribute '__univer'
print(a.univer)
```

Example Aggregation:

/55

```python
class Salary:
    def __init__(self, pay, bonus):
        self.pay = pay
        self.bonus = bonus
    def computeAnnualSalary(self):
        return (self.pay * 12) + self.bonus

class Employee:
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary
    def total_sal(self):
        # Use func of Salary: "have/has" a relationship
        return self.salary.computeAnnualSalary() 
    

salary = Salary(10000, 1500)
emp = Employee("AIVN", 25, salary)
print(emp.total_sal())
```

Example Composition:

/57

```python
class Salary:
    def __init__(self, pay, bonus):
        self.pay = pay
        self.bonus = bonus
    def computeAnnualSalary(self):
        return (self.pay * 12) + self.bonus

class Employee:
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        # Composition: one obj is part of another obj
        self.salary = salary
    def total_sal(self):
        return self.salary.computeAnnualSalary() 
    

salary = Salary(10000, 1500)
emp = Employee("AIVN", 25, salary)
print(emp.total_sal())
```

```python
class Date:
    def __init__(self, day, month, year):
        self.__day = day
        self.__month = month
        self.__year = year
    def getDay(self):
        return self.__day
    def getMonth(self):
        return self.__month
    def getYear(self):
        return self.__year
    # Aggregation       -->
    def describe(self):
        print(f'{self.__day}/{self.__month}/{self.__year}')
        
class Person:
    def __init__(self, name, dateOfBirth):
        self.__name = name
        self.__dateOfBirth = dateOfBirth
    def describe(self):
        print(self.__name)
        # Aggregation   <--
        self.__dateOfBirth.describe()


date = Date(10, 1, 2000)
peter = Person("Peter", date)
peter.describe()
```
