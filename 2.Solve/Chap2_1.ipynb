Bài 2. Cho chuỗi biểu diễn địa chỉ email:   " minhnhutvh@gmail.com"
- Rút trích và hiển thị chuỗi "gmail.com" (Đây chính là tên Host)

```python
original_str = " minhnhutvh@gmail.com"
position = original_str.find('@')
host_str = original_str[position+1 :]

print("Host name:", host_str)
print("Length:", len(host_str))
```

Bài 3. Cho chuỗi biểu diễn:  
        ‘minhnhutvh@gmail.com Sat Jan 5 09:14:16’
- Rút trích và hiển thị chuỗi “gmail.com”

```python
original_str = "minhnhutvh@gmail.com Sat Jan 5 09:14:16"

host_start_position = original_str.find('@') + 1
host_end_position = original_str.find(' ', host_start_position)

host_str = original_str[host_start_position : host_end_position]
print("Host name:", host_str)
print("Length:", len(host_str))
```

Bài 15: Năng lượng Q cần để thay đổi ΔT nhiệt độ (xét trường hợp ΔT > 0) của M gram nước được tính bằng: Q = M × C × ΔT, trong đó ΔT có đơn vị là độ (độ Celsius), M có đơn vị gram và Q có đơn vị là Joules. 

Viết chương trình theo yêu cầu:

- Đọc khối lượng của nước và sự thay đổi ΔT nhiệt độ từ người dùng. Chương trình sẽ hiển thị tổng năng lượng của nước khi thay đổi nhiệt độ mong muốn. 

- Gợi ý: Nhiệt dung riêng của nước là 4.186 J/g◦C.

- Tính toán chi phí khi thay đổi ΔT nhiệt độ của nước, biết rằng: chi phí là 8,9 cent mỗi kilowatt giờ, 1 Joules = 2.777e-7 kWh.

Hãy tính chi phí để nước đạt nhiệt độ như yêu cầu.

```python
m = int(input("Input mass of water(g): "))
dt = int(input("Input ΔT: "))
c = 4.186

q = m * c * dt
p = q * 2.777 * 0.0000001
cent = round(p * 8.9, 2)
dollar = int(cent // 100)

print("m =", m, "c =", c, "ΔT =", dt)
print("Q =", q)
print("Power(kWh):", p)
print("Cent:", cent)
print("Dollar:", dollar)
print(f"All Cost: {cent/100}")
```

Bài 16: Viết một chương trình xác định tốc độ di chuyển của một vật khi nó chạm đất. 

Người dùng sẽ nhập chiều cao mà từ đó vật đó được thả (đơn vị tính mét (m)).

Tốc độ ban đầu của một vật là 0 m/s,  gia tốc trọng trường là 9,8 $m/s^2$.


```python
from math import sqrt

h = float(input("Input height(m): "))
g = 9.8

v_f = sqrt(2 * g * h)

print("h =", h, "g =", g)
print("Velocity ground =", v_f)
```

Bài 17: Khi gió thổi trong thời tiết lạnh, làm cho không khí lạnh hơn so với thực tế. Nguyên nhân là do sự chuyển động của không khí làm tăng tốc độ làm mát cho các vật thể (chẳng hạn, như con người).

- Năm 2001, Canada, Vương quốc Anh và Hoa Kỳ đã áp dụng công thức sau đây để tính toán chỉ số gió lạnh: WCI = 13,12 + 0,6215 $T_a$ - 11,37 $V^{0.16}$ + 0,3965 $T_a V^{0.16}$

- Trong công thức $T_a$ là nhiệt độ không khí tính bằng độ C và V là tốc độ gió tính bằng km/giờ.

- Viết chương trình đọc nhiệt độ không khí và tốc độ gió từ người dùng. Hiển thị chỉ số gió lạnh được làm tròn đến số nguyên gần nhất.

```python
air_temp = float(input("Input air temparature(◦C): "))
air_speed = float(input("Input air speed(km/h): "))

wci = 13.12 + 0.6215 * air_temp \
    - 11.37 * air_speed ** 0.16 + 0.3965 * air_temp * air_speed ** 0.16

print(f"Air temparature = {air_temp}◦C, Air speed = {air_speed}km/h")
print("WCI =", round(wci, 1))
```
