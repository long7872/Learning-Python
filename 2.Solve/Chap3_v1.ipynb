````python
from math import sqrt

class SoNguyenTo:
    def __init__(self, x):
        if self.isSoNguyenTo(x):
            self.__a = x
        else: print(x, "không phải là số nguyên tố, không lưu trữ")
    
    @property
    def a(self):
        return self.__a
    @a.setter
    def a(self, a):
        if self.isSoNguyenTo(a): self.__a = a
        else: print("Không set")
    
    @staticmethod
    def isSoNguyenTo(x):
        if x == 1: return False
        for i in range(2, int(sqrt(x)) + 1):
            if x % i == 0: return False
        return True

    def timSoNguyenTo(self):
        it = self.a + 1
        while True:
            if self.isSoNguyenTo(it): 
                return it
            else: it += 1

prime1 = SoNguyenTo(1)
prime2 = SoNguyenTo(113)
x = 9
if SoNguyenTo.isSoNguyenTo(x): print(x, "is prime number")
else: print(x, "is not prime number")
print(prime2.timSoNguyenTo())
prime2.a = 126
prime2.a = 127
print(prime2.timSoNguyenTo())
````

Bài tập 6:
Xây dựng lớp tam giác (Triangle) có các thành phần:
* Các thuộc tính: 3 cạnh a, b, c của tam giác.

* Các phương thức:
    - Nhập độ dài 3 cạnh
    - Xác định xem 3 cạnh a b c có phải 3 cạnh của tam giác không, return True nếu đúng, return False nếu sai
    - Xác định kiểu tam giác
    - Tính chu vi tam giác
    - Tính diện tích tam giác

````python
from math import sqrt

class Triangle:
    def __init__(self, a=0.0, b=0.0, c=0.0):
        self.a = a
        self.b = b
        self.c = c
    
    def inputInfo(self):
        while True:
            try:
                a, b, c = map(float, input("Input a, b, c (3.0 4.0 5.0): ").split())
                if any(x <= 0 for x in [a, b, c]):
                    raise ValueError
                break
            except ValueError:
                print("Input valid float value")
        self.a, self.b, self.c = a, b, c
    
    def printInfo(self):
        print("a =", self.a)
        print("b =", self.b)
        print("c =", self.c)
        
    def is_triangle(self):
        if self.a + self.b > self.c and \
                self.a + self.c > self.b and \
                self.b + self.c > self.a:
            return True
        else: return False
    
    def get_triangle_type(self):
        if not self.is_triangle(): return "Not Triangle"
        
        x, y, z = sorted([self.a, self.b, self.c])
        
        if x == y == z: return "Equivalent Triangle"
        if x**2 + y**2 == z**2:
            if x==y or y==z or x==z:
                return "Right Isosceles Triangle"
            else: return "Isosceles Triangle"
        
        if x==y or y==z or x==z:
            return "Isosceles Triangle"
        
        return "Normal Triangle"
    
    def cal_perimeter(self):
        if not self.is_triangle(): 
            print("Edge not valid")
            return -1
        return self.a + self.b + self.c
    
    def cal_area(self):
        if not self.is_triangle(): 
            print("Edge not valid")
            return -1
        p = self.cal_perimeter() / 2
        return sqrt(p*(p-self.a)*(p-self.b)*(p-self.c))
    
tri = Triangle()
tri.inputInfo()
tri.printInfo()
print(tri.get_triangle_type())
print(tri.cal_perimeter())
print(tri.cal_area())
````

Bài tập 8:
Xây dựng lớp Phân số (Fraction) có các thành phần sau:
* Các thuộc tính: Tử số, mẫu số.
* Phương thức tạo để khởi tạo giá trị cho tử số và mẫu số.
* Các phương thức setter và getter.
* Các phương thức:
    - Nhập phân số
    - In Phân số
    - Rút gọn phân số
    - Nghịch đảo phân số
    - add(), sub(), mul(), div() tương ứng để thực hiện cộng, trừ, nhân, chia hai phân số cho nhau.

````python
class Fraction:
    def __init__(self, numerator=0, denominator=0):
        self.__a = numerator
        self.__b = denominator
    
    @property
    def numerator(self):
        return self.__a
    @numerator.setter
    def numerator(self, new_value):
        self.__a = new_value
    @property
    def denominator(self):
        return self.__b
    @denominator.setter
    def denominator(self, new_value):
        self.__b = new_value
        
    def inputInfO(self):
        while True:
            try: 
                a, b = map(int, input("Input numerator, denominator (n/d)").split('/'))
                if b == 0:
                    raise ValueError
                break
            except ValueError:
                print("numerator and denominator must be integer and denominator not 0")
        self.numerator = a
        self.denominator = b
    
    def printInfo(self):
        print(f"Fraction: {self.numerator}/{self.denominator}")
    
    @staticmethod
    def gcd(a, b):
        if b == 0: return a
        return Fraction.gcd(b, a % b)
        
    def simplify(self):
        a = self.numerator
        b = self.denominator
        gcd = Fraction.gcd(a, b)
        self.numerator = a // gcd
        self.denominator = b // gcd
    
    def inverse(self):
        temp = self.numerator
        self.numerator = self.denominator
        self.denominator = temp
    
    @staticmethod
    def add(f1: 'Fraction', f2: 'Fraction'):
        numerator = f1.numerator * f1.denominator + f2.numerator * f2.denominator
        denominator = f1.denominator * f2.denominator
        result = Fraction(numerator, denominator)
        return result

    @staticmethod
    def sub(f1: 'Fraction', f2: 'Fraction'):
        numerator = f1.numerator * f1.denominator - f2.numerator * f2.denominator
        denominator = f1.denominator * f2.denominator
        result = Fraction(numerator, denominator)
        return result
    
    @staticmethod
    def mul(f1: 'Fraction', f2: 'Fraction'):
        numerator = f1.numerator * f2.numerator
        denominator = f1.denominator * f2.denominator
        result = Fraction(numerator, denominator)
        return result
    
    @staticmethod
    def div(f1: 'Fraction', f2: 'Fraction'):
        numerator = f1.numerator * f2.denominator
        denominator = f1.denominator * f2.numerator
        result = Fraction(numerator, denominator)
        return result
    
f1 = Fraction()
f2 = Fraction()
f1.inputInfO()
f2.inputInfO()

f1.printInfo()
f2.printInfo()

f1.simplify()
f2.simplify()

f1.printInfo()
f2.printInfo()

print()

add = Fraction.add(f1, f2)
sub = Fraction.sub(f1, f2)
mul = Fraction.mul(f1, f2)
div = Fraction.div(f1, f2)

add.printInfo()
sub.printInfo()
mul.printInfo()
div.printInfo()

add.simplify()
sub.simplify()
mul.simplify()
div.simplify()

add.printInfo()
sub.printInfo()
mul.printInfo()
div.printInfo()
````

````python
class ComplexNumber:
    def __init__(self, r=0, i=0):
        self.real = r
        self.imagine = i
        
    def printInfo(self):
        operator = ""
        if self.imagine < 0: operator = "-"
        else: operator = "+"
        print(f"{self.real} {operator} {abs(self.imagine)}i")
    
    def __add__(self, other: 'ComplexNumber'):
        real = self.real + other.real
        img = self.imagine + other.real
        return ComplexNumber(real, img)
    
    def __sub__(self, other: 'ComplexNumber'):
        real = self.real - other.real
        img = self.imagine - other.real
        return ComplexNumber(real, img)
    
    def __mul__(self, other: 'ComplexNumber'):
        real = self.real * self.imagine - other.real * other.imagine
        img = self.real * other.imagine + self.imagine * other.real
        return ComplexNumber(real, img)
    
    def __truediv__(self, other: 'ComplexNumber'):
        real = (self.real * other.real + self.imagine * other.imagine) / (other.real**2 + other.imagine**2)
        img = (self.imagine * other.real - self.real * other.imagine) / (other.real**2 + other.imagine**2)
        return ComplexNumber(real, img)
    
n1 = ComplexNumber(2, 3)
n2 = ComplexNumber(4, -1)
n1.printInfo()
n2.printInfo()

(n1 + n2).printInfo()
(n1 - n2).printInfo()
(n1 * n2).printInfo()
(n1 / n2).printInfo()
````

Xây dựng lớp Vectơ có các thành phần sau:

+ Các phương thức:
    - Kiểm tra hai vectơ có bằng nhau không?
    - Tính góc giữa 2 vectơ
    - Tính module của 2 vectơ
    - Kiểm tra hai vectơ có cùng phương không?
    - Cộng hai vectơ
    - Trư hai vectơ
    - Nhân hai vectơ

Nhập vào 2 vectơ và thực hiện các phép toán trên hai vectơ đó.

````python
from math import sqrt, acos

class CustomVector:
    def __init__(self, x1=0, y1=0, x2=0, y2=0):
        self.x1 = x1
        self.y1 = y1
        self.x2 = x2
        self.y2 = y2
    
    def components(self):
        return (self.x2 - self.x1, self.y2 - self.y1)
    
    # Module
    def get_length(self):
        dx, dy = self.components()
        return sqrt(dx**2 + dy**2)
            
    def is_equal(self, other: 'CustomVector'):
        return self.components() == other.components()
    
    def get_angle(self, other: 'CustomVector'):
        dx1, dy1 = self.components()
        dx2, dy2 = other.components()
        
        dot = dx1*dx2 + dy1*dy2
        len_product = self.get_length() * other.get_length()
        
        if len_product == 0:
            return None
        
        return(acos(dot / len_product))
    
    def is_parallel(self, other):
        dx1, dy1 = self.components()
        dx2, dy2 = other.components()
        return dx1*dy2 == dy1*dx2
    
    def __add__(self, other: 'CustomVector'):
        dx1, dy1 = self.components()
        dx2, dy2 = other.components()
        return CustomVector(0, 0, dx1+dx2, dy1+dy2)
    
    def __sub__(self, other: 'CustomVector'):
        dx1, dy1 = self.components()
        dx2, dy2 = other.components()
        return CustomVector(0, 0, dx1-dx2, dy1-dy2)
    
    def __mul__(self, other: 'CustomVector'):
        dx1, dy1 = self.components()
        dx2, dy2 = other.components()
        return dx1*dx2 + dy1*dy2
    
    def __str__(self):
        dx, dy = self.components()
        return f"Vector({dx}, {dy})"
    
v1 = CustomVector(0,0,3,4)
v2 = CustomVector(0,0,6,8)

print("Module v1:", v1.get_length())
print("Equal:", v1.is_equal(v2))
print("Parallel:", v1.is_parallel(v2))
print("Angle:", v1.get_angle(v2))

print("Add:", v1 + v2)
print("Sub:", v1 - v2)
print("Dot:", v1 * v2)
````

Xây dựng lớp đa thức (polylomial) và các phép toán trên đa thức. Thực hiện nhập vào 2 đa thức và tính tổng, tích của nó.

````python
from IPython.display import display, Math

class Polynomial:
    def __init__(self, terms=None):
        self.terms=terms if terms else {}
    
    @staticmethod
    def normalize_key(key):
        return tuple(sorted([(v,p) for v,p in key if p!=0]))
    
    def add_term(self, coeff, powers):
        key = Polynomial.normalize_key(powers)
        self.terms[key] = self.terms.get(key,0) + coeff
        if self.terms[key] == 0:
            del self.terms[key]

    def __add__(self, other: 'Polynomial'):
        result = Polynomial(self.terms.copy())
        for k,v in other.terms.items():
            result.terms[k] = result.terms.get(k,0) + v
            if result.terms[k] == 0:
                del result.terms[k]
        
        return result
    
    def __sub__(self, other: 'Polynomial'):
        result = Polynomial(self.terms.copy())
        for k,v in other.terms.items():
            result.terms[k] = result.terms.get(k,0) - v
            if result.terms[k] == 0:
                del result.terms[k]
        
        return result
    
    def __mul__(self, other: 'Polynomial'):
        result = Polynomial()
        
        for k1,c1 in self.terms.items():
            for k2, c2 in other.terms.items():
                power_map = {}
                for v,p in k1:
                    power_map[v] = power_map.get(v,0)+p
                for v,p in k2:
                    power_map[v] = power_map.get(v,0)+p
                new_key = Polynomial.normalize_key(power_map.items())
                result.terms[new_key] = result.terms.get(new_key,0)+c1*c2
        
        return result
    
    def evaluate(self, **values):
        total = 0
        for key, coeff in self.terms.items():
            term = coeff
            for var, pow_ in key:
                term = term * values.get(var,0)**pow_
            total = total + term
        return total
    
    def __str__(self):
        if not self.terms:
            return '0'
        
        res = []
        for key, coeff in self.terms.items():
            term = f'{coeff}'
            for v,p in key:
                term = term + f'{v}^{p}'
            res.append(term)
        
        return ' + '.join(res)
    
    def _repr_latex_(self):
        if not self.terms:
            return '$0$'
        
        res = []
        for key, coeff in self.terms.items():
            term = ""
            
            if coeff == -1 and key:
                term += '-'
            elif coeff != 1 or not key:
                term += str(coeff)
            
            for v,p in key:
                term += v
                if p != 1:
                    term += f"^{{{p}}}"
            
            res.append(term)
        
        latex = ' + '.join(res).replace("+ -", '- ')
        return f'${latex}$'
    
    def to_latex(self):
        if not self.terms:
            return "0"

        res = []
        for key, coeff in self.terms.items():
            term = ""

            if coeff == -1 and key:
                term += "-"
            elif coeff != 1 or not key:
                term += str(coeff)

            for v,p in key:
                term += v
                if p != 1:
                    term += f"^{{{p}}}"

            res.append(term)

        return ' + '.join(res).replace("+ -", "- ")
    
p1 = Polynomial()
p1.add_term(1,[('x',1),('y',1)])
p1.add_term(1,[('x',2)])

p2 = Polynomial()
p2.add_term(1,[('y',1),('z',1)])

print(p1)
print(p2)

display(p1)
display(p2)

display(Math("Sum = " + (p1+p2).to_latex()))
display(Math("Product = " + (p1*p2).to_latex()))
````

Bài tập 12:

Xây dựng lớp ma trận và các phép toán trên ma trận. Thực hiện nhập vào 2 ma trận và tính tổng, tích của nó.

````python
class Matrix:
    def __init__(self, rows=0, cols=0):
        self.rows=rows
        self.cols=cols
        self.data=[[0]*cols for _ in range(rows)]
    
    def inputData(self, st: str):
        rows = st.split(', ')
        matrix = []
        
        for row in rows:
            row = list(map(float,row.split(' ')))
            matrix.append(row)
        
        col_len = len(matrix[0])
        for row in matrix:
            if len(row)!=col_len:
                raise ValueError("Invalid matrix format")
        
        self.data = matrix
        self.rows = len(matrix)
        self.cols = col_len
    
    def printInfo(self):
        for row in self.data:
            print(' '.join(map(str,row)))
        
    def __add__(self, other: 'Matrix'):
        if self.rows!=other.rows or self.cols!=other.cols:
            raise ValueError("Matrix size mismatch for addition")
        
        result = Matrix(self.rows, self.cols)
        
        for i in range(self.rows):
            for j in range(self.cols):
                result.data[i][j] = self.data[i][j] + other.data[i][j]
        
        return result
    
    def __mul__(self, other: 'Matrix'):
        if self.cols!=other.rows:
            raise ValueError("Matrix size mismatch for multiplication")
        
        result = Matrix(self.rows, other.cols)
        
        for i in range(self.rows):
            for j in range(other.cols):
                for k in range(self.cols):
                    result.data[i][j] += self.data[i][k] * other.data[k][j]
                    
        return result
        
m1 = Matrix()
m2 = Matrix()
m3 = Matrix()

m1.inputData("1 2 3, 4 5 6")
m2.inputData("7 8, 9 10, 11 12")
m3.inputData("3.4 2.1 5.7, 8.7 3.5 8.5")

print("Matrix 1:")
m1.printInfo()

print("\nMatrix 2:")
m2.printInfo()

print("\nMatrix 3:")
m3.printInfo()

try:
    print("\nSum 1+2:")
    (m1+m2).printInfo()
except:
    print("Cannot add")

try:
    print("\nSum 1+3:")
    (m1+m3).printInfo()
except:
    print("Cannot add")
    
try:
    print("\nProduct 1*2:")
    (m1*m2).printInfo()
except:
    print("Cannot multiply")
````

Bài tập 13:

Xây dựng lớp đa giác, hình bình hành thừa kế từ đa giác, hình chữ nhật thừa kế từ hình bình hành và hình vuông thừa kế từ hình chữ nhật. Nhập vào các thuộc tính cần thiết của mỗi hình và tính chu vi, diện tích của hình đó.

````python
from abc import ABC, abstractmethod

class Polygon(ABC):
    @abstractmethod
    def perimeter(self):
        pass
    
    @abstractmethod
    def area(self):
        pass
    
class Parallelogram(Polygon):
    def __init__(self, base, side, height):
        self.base = base
        self.side = side
        self.height = height
    
    def perimeter(self):
        return 2 * (self.base + self.side)
    
    def area(self):
        return self.base * self.height
        
class Rectangle(Parallelogram):
    def __init__(self, length, width):
        super().__init__(length, width, width)

class Square(Rectangle):
    def __init__(self, side):
        super().__init__(side, side)
        self.side = side

p = Parallelogram(5, 3, 4)
print("Parallelogram perimeter:", p.perimeter())
print("Parallelogram area:", p.area())

r = Rectangle(6, 4)
print("Rectangle perimeter:", r.perimeter())
print("Rectangle area:", r.area())

s = Square(5)
print("Square perimeter:", s.perimeter())
print("Square area:", s.area())
````

Xây dựng lớp điểm, lớp elip thừa kế từ lớp điểm, lớp đường tròn thừa kế từ lớp elip. Nhập vào các thuộc tính cần thiết của elip và tính diện tích.

````python
from math import pi

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
class Ellipse(Point):
    def __init__(self, x, y, a, b):
        super().__init__(x, y)
        self.a = a
        self.b = b
    
    def area(self):
        return pi * self.a * self.b

class Circle(Ellipse):
    def __init__(self, x, y, r):
        super().__init__(x, y, r, r)
        self.r = r

p = Point(2, 3)
print(f"Point: {p.x}, {p.y}")
e = Ellipse(p.x, p.y, 5, 4)
print(f"Axis: {e.a}, {e.b}. Area: {e.area()}")
c = Circle(p.x, p.y, 4)
print(f"Radius: {c.r}. Area: {c.area()}")
````

Bài tập 15:

Xây dựng lớp tam giác, lớp tam giác vuông, tam giác cân thừa kế từ lớp tam giác. Lớp tam giác đều thừa kế từ lớp tam giác cân.

````python
from math import sqrt

class Triangle:
    def __init__(self, a, b, c):
        if not self.is_valid(a, b, c):
            raise ValueError("Invalid triangle")
        self.a = a
        self.b = b
        self.c = c
        
    def is_valid(self, a, b, c):
        return a + b > c and b + c > a and a + c > b
    
    def perimeter(self):
        return self.a + self.b + self.c
    
    def area(self):
        # Heron
        p = self.perimeter() / 2
        return sqrt(p*(p-self.a)*(p-self.b)*(p-self.c))

class RightTriangle(Triangle):
    def __init__(self, a, b, c):
        super().__init__(a, b, c)
        if not self.is_right(a, b, c):
            raise ValueError("Not right triangle")
    
    def is_right(self, a, b, c):
        return a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2

class IsoscelesTriangle(Triangle):
    def __init__(self, a, b, c):
        super().__init__(a, b, c)
        if not self.is_isosceles(a, b, c):
            raise ValueError("Not isosceles triangle")
    
    def is_isosceles(self, a, b, c):
        return a == b or b == c or a == c
    
class EquilateralTriangle(IsoscelesTriangle):
    def __init__(self, a):
        super().__init__(a, a, a)
        
t = Triangle(3,4,5)
print("Triangle area:", t.area())

rt = RightTriangle(3,4,5)
print("Right area:", rt.area())

iso = IsoscelesTriangle(5,5,6)
print("Isosceles area:", iso.area())

eq = EquilateralTriangle(4)
print("Equilateral area:", eq.area())
````

````python
from math import sqrt

class MyPoint:
    def __init__(self, x=0, y=0):
        self.__x = x
        self.__y = y
    
    @property
    def x(self):
        return self.__x
    @x.setter
    def x(self, new_value):
        self.__x = new_value
    @property
    def y(self):
        return self.__y
    @y.setter
    def y(self, new_value):
        self.__y = new_value
    
    def setXY(self, x, y):
        self.__x = x
        self.__y = y
        
    def toString(self):
        return f"({self.x}, {self.y})"
    
    def distanceXY(self, x, y):
        dx = x - self.x
        dy = y - self.y
        return sqrt(dx**2 + dy**2)
    
    def distancePoint(self, point1: 'MyPoint'):
        dx = point1.x - self.x
        dy = point1.y - self.y
        return sqrt(dx**2 + dy**2)
    
class TestMyPoint:
    @staticmethod
    def main():
        p1 = MyPoint(3, 0)
        p2 = MyPoint(0, 4)
        print(p1.toString())
        print(p2.toString())
        
        print(p1.distancePoint(p2))
        print(p1.distanceXY(5,6))
        
TestMyPoint.main()
````

````python
import random as r

class BankAccount:
    def __init__(self, accountNumber, accountName):
        self.__accountNumber = accountNumber
        self.__accountName = accountName
        self.__balance = 0
    
    def getAccountNumber(self):
        return self.__accountNumber
    
    def getAccountName(self):
        return self.__accountName
    
    def getBalance(self):
        return self.__balance
    
    def deposit(self, money):
        self.__balance += money
        return True
    
    def withdraw(self, money):
        if self.__balance >= money:
            self.__balance -= money
            return True
        else: return False

class Bank:
    def __init__(self, bankAccountList: list[BankAccount] = []):
        self.bankAccountList = bankAccountList
    
    def search(self, accountNumber):
        for i, acc in enumerate(self.bankAccountList):
            if acc.getAccountNumber() == accountNumber:
                return i
        return -1

    def getTotal(self):
        return len(self.bankAccountList)
    
    def getItem(self, accountNumber):
        for acc in self.bankAccountList:
            if acc.getAccountNumber() == accountNumber:
                return acc
        return None
    
    def addAccount(self, accountNumber, accountName):
        if self.search(accountNumber) != -1:
            return False
        new_account = BankAccount(accountNumber, accountName)
        self.bankAccountList.append(new_account)
        return True
    
    def depositMoney(self, accountNumber, money):
        accountIndex = self.search(accountNumber)
        if accountIndex == -1:
            return False
        else: 
            account = self.bankAccountList[accountIndex]
            account.deposit(money)
            return True
    
    def withdrawMoney(self, accountNumber, money):
        accountIndex = self.search(accountNumber)
        if accountIndex == -1:
            return False
        else: 
            account = self.bankAccountList[accountIndex]
            if account.withdraw(money): 
                return True
            else: return False
    
    def removeAccount(self, accountNumber):
        accountIndex = self.search(accountNumber)
        if accountIndex == -1:
            return False
        else: 
            self.bankAccountList.pop(accountIndex)
            return True
        
bank = Bank()
ls = [(1050008855, "A"), 
      (1050008854, "B"), 
      (1050008853, "C"), 
      (1050008852, "D"), 
      (1050008851, "E"), 
      (1050008850, "F"), 
      ]
print("Add bank account list: \n")
for acc in ls:
    print("Attempt add:", acc[0])
    is_done = bank.addAccount(acc[0], acc[1])
    if is_done: print(f"{acc[0]}:{acc[1]} add successfully")
    else: print(f"{acc[0]}:{acc[1]} add failed")

print("\n\nGenerate money for account list\n")
for acc in bank.bankAccountList:
    money = r.randint(0, 10000000000)
    print("Attempt generate money for:", acc.getAccountNumber())
    is_done = acc.deposit(money)
    if is_done: print(f"{acc.getAccountNumber()}:{acc.getAccountName()} add {money} successfully")
    else: print(f"{acc.getAccountNumber()}:{acc.getAccountName()} add {money} failed")
    
print("\n\nPrint infomation\n")
count = 1
for acc in bank.bankAccountList:
    print(f"{count}. {acc.getAccountNumber()}:{acc.getAccountName()} has {acc.getBalance()} VND")
    count += 1
count = 0

print("\n\nDeposit 500M VND to 1050008855\n")
print("Balance:", bank.getItem(1050008855).getBalance())
is_done = bank.depositMoney(1050008855, 500000000)
if is_done: print(f"1050008855 deposit 500000000 successfully")
else: print(f"1050008855 deposit 500000000 failed")
print("Balance:", bank.getItem(1050008855).getBalance())

print("\n\nWithdraw 30M VND to 1050008854\n")
print("Balance:", bank.getItem(1050008854).getBalance())
is_done = bank.withdrawMoney(1050008854, 30000000)
if is_done: print(f"1050008854 deposit 30000000 successfully")
else: print(f"1050008854 deposit 30000000 failed")
print("Balance:", bank.getItem(1050008854).getBalance())
````

Sách được đặt trên từng kệ sách (mỗi kệ sách bao gồm nhiều cuốn sách). Mỗi kệ sách có 1 số thứ tự riêng.

Các kệ sách được phân ra theo đầu mục sách như: Công nghệ thông tin, truyện ngắn, sách giáo khoa, tiểu thuyết...

Mỗi quyển sách lại được đánh một mã khác nhau.

Trên phần mềm có lưu trữ thêm thông tin về tác giả, năm xuất bản, giá tiền của sách

Để tiện cho việc quản lý. Mỗi khi có một cuốn sách mới, người thủ thư sẽ đưa cuốn sách lên kệ theo đúng đầu mục, ngoài ra, thủ thư sẽ đặt cuốn sách sao cho trong 1 kệ, các cuốn sách xếp theo vần alphabet (sắp xếp từ A-->Z)

Một đặc trưng nữa của phần mềm, đấy là có thể lấy ra cuốn sách có giá cao nhất và cuốn sách có giá thấp nhất theo từng kệ sách, chỉ cần nhập vào số thứ tự kệ sách.

Để tìm kiếm đúng cuốn sách cần thiết, người dùng chỉ cần nhập vào mã sách là phần mềm sẽ trả về thông tin cuốn sách tương ứng.

Hãy viết phần mềm quản lý thư viện theo mô tả như trên.

````python
class Book:
    def __init__(self, book_id="", name="", author="", pub_year="", price=0.0):
        self.book_id=book_id
        self.name=name
        self.author=author
        self.pub_year=pub_year
        self.price=price
        
    def is_price_valid(self, price):
        if isinstance(price, float) and price>0:
            return True
        else: return False
    
    def inputInfo(self):
        print("Input infomation to create Book on System: ")
        self.book_id=input("Book ID: ")
        self.name=input("Book name: ")
        self.author=input("Author: ")
        self.pub_year=input("Publication year: ")
        while True:
            try:
                price=float(input("Price: "))
                if self.is_price_valid(price):
                    self.price=price
                    break
                else: print("Price value not correctly")
            except ValueError:
                print("Price value must be float")
                
    def showInfo(self):
        print("Infomation of Book: ")
        print("Book ID:", self.book_id)
        print("Book name:", self.name)
        print("Author:", self.author)
        print("Publication year:", self.author)
        print("Price:", self.price)

class Bookshelf:
    def __init__(self, type=""):
        self.type=type
        self.book_list: list[Book] = []
    
    def is_exist(self, book_id):
        for book in self.book_list:
            if book.book_id == book_id:
                return book
        return None
    
    def addBook(self, book: Book):
        if self.is_exist(book.book_id) is None:
            self.book_list.append(book)
            self.sort_book_AZ()
            return True
        else: return False
    
    def removeBook(self, book_id):
        book = self.is_exist(book_id)
        if book is not None:
            self.book_list.remove(book)
            return True
        else: return False
    
    def show_books(self):
        print(f"Show all books on this {self.type} bookshelf")
        for i, book in enumerate(self.book_list):
            print(f"Book {i}. Book ID: {book.book_id} - Book name: {book.name} - Author: {book.author} - Publication year: {book.pub_year} - Price: {book.price}")
    
    def sort_book_AZ(self):
        self.book_list=sorted(self.book_list, key=lambda x: x.name)
        
    def getHighestAndLowerPriceBook(self):
        if not self.book_list:
            return False
        min_book=self.book_list[0]
        max_book=self.book_list[0]
        for book in self.book_list:
            if book.price > max_book.price:
                max_book = book
            if book.price < min_book.price:
                min_book = book
        return (max_book, min_book)

class Library:
    def __init__(self):
        shelf_list = []
        shelf_types = ["Công nghệ thông tin", "Truyện ngắn", "Sách giáo khoa", "Tiểu thuyết"]
        for type in shelf_types:
            shelf_list.append(Bookshelf(type))
        self.shelf_list: list[Bookshelf]=shelf_list
    
    def showInfo(self):
        print("Show all bookshelf already have: ")
        for i, shelf in enumerate(self.shelf_list):
            print(f"Shelf {i}. Type: {shelf.type}")
    
    def is_exist(self, shelf_type):
        for shelf in self.shelf_list:
            if shelf.type == shelf_type:
                return shelf
        return None
    
    def addBookshelf(self, bookshelf: Bookshelf):
        if self.is_exist(bookshelf.type) is None:
            self.shelf_list.append(bookshelf)
            return True
        else: return False
    
    def removeBookshelf(self, shelf_type):
        shelf = self.is_exist(shelf_type)
        if shelf is not None:
            self.shelf_list.remove(shelf)
            return True
        else: return False
    
    def addBookToShelf(self, shelf_type, book: Book):
        shelf = self.is_exist(shelf_type)
        if shelf is not None:
            if shelf.addBook(book):
                return True
        return False
    
    def removeBookFromShelf(self, shelf_type, book_id):
        shelf = self.is_exist(shelf_type)
        if shelf is not None:
            if shelf.removeBook(book_id):
                return True
        return False
    
    def getHighestAndLowestPriceInShelf(self, shelf_type):
        shelf = self.is_exist(shelf_type)
        if shelf is not None:
            max_price_book, min_price_book = shelf.getHighestAndLowerPriceBook()
            return (max_price_book, min_price_book)
        return None
    
    def getBookByID(self, book_id):
        for shelf in self.shelf_list:
            for book in shelf.book_list:
                if book.book_id == book_id:
                    return book
        return None

    @staticmethod
    def main():
        library = Library()
        while True: 
            print(f"\n{'='*10} Menu {'='*10}")
            print("Chọn 1: Show all bookshelf")
            print("Chọn 2: Add bookshelf")
            print("Chọn 3: Remove bookshelf")
            print("Chọn 4: Add Book")
            print("Chọn 5: Remove Book")
            print("Chọn 6: Show highest and lowest price of book")
            print("Chọn 7: Find book by Book ID")
            print("Chọn 8: Thoát\n")
            
            mode = 0
            while True: 
                try:
                    lower_bound = 1
                    upper_bound = 8
                    mode = int(input(f"Choose mode ({lower_bound}-{upper_bound}): "))
                    if mode in range(lower_bound, upper_bound+1):
                        break
                    else: print(f"Choose between {lower_bound} and {upper_bound}")
                except ValueError:
                    print("Input valid integer number")

            match mode:
                case 1:
                    print()
                    library.showInfo()
                    print()
                case 2:
                    print()
                    book_type = input("Input book type you want to add to library: ")
                    shelf = Bookshelf(book_type)
                    if library.addBookshelf(shelf):
                        print(f"Add {book_type} bookshelf successfully")
                    else: print(f"Add {book_type} bookshelf failed")
                    print()
                case 3:
                    print()
                    book_type = input("Input book type you want to remove from library: ")
                    if library.removeBookshelf(book_type):
                        print(f"Remove {book_type} bookshelf successfully")
                    else: print(f"Remove {book_type} bookshelf failed")
                    print()
                case 4:
                    print()
                    shelf_type = input("Input shelf type you want to add book inside: ")
                    shelf = library.is_exist(shelf_type)
                    if shelf is not None:
                        print("Bookshelf has this type detected! Continue with input book properties...")
                        book = Book()
                        book.inputInfo()
                        shelf.addBook(book)
                    else: print("Your bookshelf type has not exist! Please input correctly or create new one.")
                    print()
                case 5:
                    print()
                    shelf_type = input("Input shelf type you want to remove book inside: ")
                    shelf = library.is_exist(shelf_type)
                    if shelf is not None:
                        print("Bookshelf has this type detected! Continue with input book properties...")
                        book_id = input("Input Book ID to remove this book from this bookshelf: ")
                        shelf.removeBook(book_id)
                    else: print("Your bookshelf type has not exist! Please input correctly or create new one.")
                    print()
                case 6:
                    print()
                    shelf_type = input("Input shelf type you want to show highest and lowest price: ")
                    shelf = library.is_exist(shelf_type)
                    if library.is_exist(shelf_type) is not None:
                        print("Bookshelf has this type detected! Continue with input book properties...")
                        max_price_book, min_price_book = shelf.getHighestAndLowerPriceBook()
                        print("\nMax Price: ")
                        max_price_book.showInfo()
                        print("\nMin Price: ")
                        min_price_book.showInfo()
                    else: print("Your bookshelf type has not exist! Please input correctly or create new one.")
                    print()
                case 7:
                    print()
                    book_id = input("Input book ID to find: ")
                    book = library.getBookByID(book_id)
                    if book is not None:
                        print(f"Book with ID: {book_id} has found!")
                        book.showInfo()
                    else: print(f"Book with ID: {book_id} not found!")
                    print()
                case 8:
                    break

Library.main()
````

````python
from abc import ABC, abstractmethod

class Polygon(ABC):
    @abstractmethod
    def perimeter(self):
        pass
    
    @abstractmethod
    def area(self):
        pass
    
class Parallelogram(Polygon):
    def __init__(self, base, side, height):
        self.base = base
        self.side = side
        self.height = height
    
    def perimeter(self):
        return 2 * (self.base + self.side)
    
    def area(self):
        return self.base * self.height
        
class Rectangle(Parallelogram):
    def __init__(self, length, width):
        super().__init__(length, width, width)

class Square(Rectangle):
    def __init__(self, side):
        super().__init__(side, side)
        self.side = side

p = Parallelogram(5, 3, 4)
print("Parallelogram perimeter:", p.perimeter())
print("Parallelogram area:", p.area())

r = Rectangle(6, 4)
print("Rectangle perimeter:", r.perimeter())
print("Rectangle area:", r.area())

s = Square(5)
print("Square perimeter:", s.perimeter())
print("Square area:", s.area())
````

Bài tập 5:

Xây dựng lớp điểm, lớp elip thừa kế từ lớp điểm, lớp đường tròn thừa kế từ lớp elip. Nhập vào các thuộc tính cần thiết của elip và tính diện tích.

````python
from math import pi

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
class Ellipse(Point):
    def __init__(self, x, y, a, b):
        super().__init__(x, y)
        self.a = a
        self.b = b
    
    def area(self):
        return pi * self.a * self.b

class Circle(Ellipse):
    def __init__(self, x, y, r):
        super().__init__(x, y, r, r)
        self.r = r

p = Point(2, 3)
print(f"Point: {p.x}, {p.y}")
e = Ellipse(p.x, p.y, 5, 4)
print(f"Axis: {e.a}, {e.b}. Area: {e.area()}")
c = Circle(p.x, p.y, 4)
print(f"Radius: {c.r}. Area: {c.area()}")
````

Bài tập 6:

Xây dựng lớp tam giác, lớp tam giác vuông, tam giác cân thừa kế từ lớp tam giác. Lớp tam giác đều thừa kế từ lớp tam giác cân.

````python
from math import sqrt

class Triangle:
    def __init__(self, a, b, c):
        if not self.is_valid(a, b, c):
            raise ValueError("Invalid triangle")
        self.a = a
        self.b = b
        self.c = c
        
    def is_valid(self, a, b, c):
        return a + b > c and b + c > a and a + c > b
    
    def perimeter(self):
        return self.a + self.b + self.c
    
    def area(self):
        # Heron
        p = self.perimeter() / 2
        return sqrt(p*(p-self.a)*(p-self.b)*(p-self.c))

class RightTriangle(Triangle):
    def __init__(self, a, b, c):
        super().__init__(a, b, c)
        if not self.is_right(a, b, c):
            raise ValueError("Not right triangle")
    
    def is_right(self, a, b, c):
        return a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2

class IsoscelesTriangle(Triangle):
    def __init__(self, a, b, c):
        super().__init__(a, b, c)
        if not self.is_isosceles(a, b, c):
            raise ValueError("Not isosceles triangle")
    
    def is_isosceles(self, a, b, c):
        return a == b or b == c or a == c
    
class EquilateralTriangle(IsoscelesTriangle):
    def __init__(self, a):
        super().__init__(a, a, a)
        
t = Triangle(3,4,5)
print("Triangle area:", t.area())

rt = RightTriangle(3,4,5)
print("Right area:", rt.area())

iso = IsoscelesTriangle(5,5,6)
print("Isosceles area:", iso.area())

eq = EquilateralTriangle(4)
print("Equilateral area:", eq.area())
````
