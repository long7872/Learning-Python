```python
mycursor = mydb.cursor()
sql = "SELECT u.name as user, p.name as favorite \
    FROM users as u INNER JOIN products as p ON u.fav = p.id"
mycursor.execute(sql)
myresult = mycursor.fetchall()
for x in myresult:
    print(x)
```

```python
mycursor = mydb.cursor()
sql = "SELECT u.name as user, p.name as favorite \
    FROM users as u LEFT JOIN products as p ON u.fav = p.id"
mycursor.execute(sql)
myresult = mycursor.fetchall()
for x in myresult:
    print(x)
```

```python
mycursor = mydb.cursor()
sql = "SELECT u.name as user, p.name as favorite \
    FROM users as u RIGHT JOIN products as p ON u.fav = p.id"
mycursor.execute(sql)
myresult = mycursor.fetchall()
for x in myresult:
    print(x)
```
