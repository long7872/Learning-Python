```python
class StringLength:
    @staticmethod
    def printLonger(st1, st2):
        if len(st1) > len(st2):
            print(st1)
        elif len(st1) < len (st2):
            print(st2)
        else: print(st1); print(st2)

st1 = "hello world"
st2 = "???????"
StringLength.printLonger(st1, st2)
```

Bài 44: Định nghĩa một hàm có thể tạo và in danh sách chứa các giá trị bình phương của các số từ 1 đến 20 (tính cả 1 và 20). Yêu cầu: Sử dụng toán tử ** để lấy giá trị bình phương; Sử dụng range() cho vòng lặp; Sử dụng list.append() để thêm giá trị vào danh sách.

```python
class CalSquare:
    ls = []
    @classmethod
    def calculateSquare(cls, lower_bound = 1, upper_bound = 20):
        cls.ls = []
        for i in range(lower_bound, upper_bound+1):
            cls.ls.append(i**2)
    @classmethod
    def printList(cls):
        print(', '.join(str(k) for k in cls.ls))

CalSquare.calculateSquare()
CalSquare.printList()
```

Bài 48: Viết hàm có 03 tham số và trả về giá trị trung bình của các tham số đó. Bao gồm một chương trình chính đọc ba giá trị từ người dùng và hiển thị giá trị trung bình của chúng.

```python
class AvgNum:
    @staticmethod
    def avg_3_num(*a):
        total = 0
        count = 0
        for item in a:
            total += item
            count += 1
        avg = total / count
        return avg

a = float(input())
b = float(input())
c = float(input())
avg = AvgNum.avg_3_num(a, b, c)
print(avg)
```

Bài 49: Số nguyên tố là một số nguyên lớn hơn 1 và chỉ chia hết cho 1 và chính nó. Viết hàm xác định tham số của nó có phải là số nguyên tố hay không, trả về True nếu đúng và False nếu không phải. Viết chương trình chính đọc số nguyên từ người dùng và hiển thị thông báo cho biết đó có phải là số nguyên tố hay không.

```python
from math import sqrt

class PrimeNum:
    @staticmethod
    def check_prime(n):
        if n == 1: return False
        for i in range(2, int(sqrt(n)) + 1):
            if n % i == 0: return False
        return True

n = int(input())
is_prime = PrimeNum.check_prime(n)
if is_prime: print(n, "is a prime number")
else: print(n, "is not a prime number")
```

Bài 50: Viết hàm không có tham số để tạo mật khẩu ngẫu nhiên. Mật khẩu phải có độ dài ngẫu nhiên từ 7 đến 10 ký tự. Mỗi ký tự phải được chọn ngẫu nhiên từ các vị trí 33 đến 126 trong bảng ASCII. Hàm sẽ trả về mật khẩu được tạo ngẫu nhiên. Hiển thị mật khẩu được tạo ngẫu nhiên trong chương trình chính.

```python
import random as r
class Password:
    @staticmethod
    def generate_password():
        length = r.randint(7, 10)
        passw = ""
        for _ in range(1, length + 1):
            c = chr(r.randint(33, 126))
            passw += c
        return passw

print(Password.generate_password())
```

Bài 51: Viết một hàm có một tham số password để xác định xem mật khẩu có tốt hay không. Một mật khẩu tốt là một mật khẩu dài ít nhất 8 ký tự và chứa ít nhất một chữ cái viết hoa, ít nhất một chữ cái viết thường và ít nhất một số. Hàm sẽ trả về True nếu mật khẩu là tốt, ngược thì nó sẽ trả về Fales. Chương trình có một chương trình chính đọc mật khẩu từ người dùng và hiển thị xem nó có tốt hay không.

```python
import random as r
class Password:
    @staticmethod
    def generate_password():
        length = r.randint(7, 10)
        passw = ""
        for _ in range(1, length + 1):
            c = chr(r.randint(33, 126))
            passw += c
        return passw
    @staticmethod
    def check_strong_password(passw: str):
        if len(passw) < 8: return False
        has_upper = False
        has_lower = False
        has_numberic = False
        for c in passw:
            if c.isupper(): has_upper = True
            elif c.islower(): has_lower= True
            elif c.isnumeric(): has_numberic = True
        if not has_lower: return False
        if not has_upper: return False
        if not has_numberic: return False
        
        return True

passw = input()
print(Password.check_strong_password(passw))
```

Yêu cầu: Hàm chứa một tham số, nếu tham số đó là một số hoàn hảo thì hàm sẽ trả về True. Nếu không nó sẽ trả về False. Ngoài ra, hãy viết một chương trình chính sử dụng hàm này để xác định và hiển thị tất cả các số hoàn hảo trong khoảng từ 1 đến 10.000

```python
class PerfectNumber:
    @staticmethod
    def check_perfect(n):
        if n == 1: return False
        sum_divisor = 1
        for i in range(2, n//2 + 1):
            if n % i == 0: sum_divisor += i
        if sum_divisor == n: return True
        else: return False
        
for i in range(1, 10000+1):
    is_perfect = PerfectNumber.check_perfect(i)
    if is_perfect: print(i, "is a perfect number")
    # else: print(n, "is not a perfect number")
```

Bài 53: Viết một hàm trả về một danh sách chứa mọi danh sách con có thể có của danh sách. Ví dụ: danh sách con của [1, 2, 3] là [], [1], [2], [3], [1, 2], [2, 3] và [1, 2, 3].

```python
class SubSet:
    def __init__(self, ls: list):
        self.ls = ls
        self.subset_list = []
    def find_all_subset(self):
        subset = []
        self.backtrack(subset=subset)
    def backtrack(self, start = 0, subset: list = []):
        self.subset_list.append(subset[:])
        for i in range(start, len(self.ls)):
            subset.append(self.ls[i])
            self.backtrack(start=i + 1, subset=subset)
            subset.pop()
    def print_all_subset(self):
        print(self.subset_list)

subset = SubSet(ls = [1,2,3,4])
subset.find_all_subset()
subset.print_all_subset()
```

- apples

- apples, oranges and bananas

Viết hàm lấy danh sách các chuỗi làm tham số của nó. Hàm sẽ trả về một chuỗi chứa tất cả các từ trong danh sách, được định dạng theo cách được mô tả như trên. Hàm hoạt động cho danh sách có độ dài bất kỳ. Code bao gồm một chương trình chính đọc một số từ do người dùng nhập vào, định dạng chúng bằng cách gọi hàm và sau đó hiển thị kết quả mà hàm trả về

```python
class FormatHuman:
    @staticmethod
    def format(word_list: list):
        num_of_word = len(word_list)
        if num_of_word == 1: return word_list[0]
        if num_of_word > 1: 
            st = ', '.join(word_list[:-1]) + " and " + word_list[-1]
            return st
        return ""

word_line = input()
word_list = word_line.split()
print(FormatHuman.format(word_list))
```

Bài 55. Tệp nén là một loại lưu trữ chứa một hoặc nhiều tệp đã bị giảm kích thước. Nén tập tin trong các hệ điều hành hiện đại thường khá đơn giản. Viết chương trình nén và giải nén tệp bằng Python

```python
import zipfile as zf

class ZipUtils:
    @staticmethod
    def compress(source_path: str, destination_path: str):
        with zf.ZipFile(destination_path, 'w', zf.ZIP_DEFLATED) as z:
            z.write(source_path)
    
    @staticmethod
    def extract(source_path: str, destination_path: str):
        with zf.ZipFile(source_path, 'r') as z:
            z.extract(destination_path)

# ZipUtils.compress("source/chap_3/1_image.png", "source/chap_3/1_image.zip")
ZipUtils.extract("source/chap_3/1_image.zip", "source/chap_3/1_image.png")
```
