Bài 1: Xử lý ngoại lệ FileNotFoundError:
- Lỗi này xuất hiện khi mở 1 file mà không tồn tại. 
- Không tồn tại có thể do file chưa được tạo hoặc đường dẫn đến file không đúng.

```python
filename = "notfound.txt"
try:
    with open(filename) as f:
        contents =  f.read()
except FileNotFoundError:
    msg = "File " + filename + " File Not Found."
    print(msg)
```

Bài 2: về xử lý ngoại lệ FileNotFoundError khi thực hiện chương trình đếm số từ trong file text

```python
filename = "source/chap_2/notfound.txt"
try:
    with open(filename) as f:
        contents =  f.read()
except FileNotFoundError:
    msg = "File " + filename + " File Not Found."
    print(msg)
else:
    words = contents.split()
    num_words = len(words)
    print("File " + filename + " have " + str(num_words) + " words.")
```

Bài 3: về thực hiện chương trình đếm số từ trong nhiều file text

```python
def count_words(filename):
    try:
        with open(filename) as f:
            contents =  f.read()
    except FileNotFoundError:
        msg = "File " + filename + " File Not Found."
        print(msg)
    else:
        words = contents.split()
        num_words = len(words)
        print("File " + filename + " have " + str(num_words) + " words.")

# filename = "source/chap_2/notfound.txt"
filenames = ['source/chap_2/notfound.txt', 'f2.txt', 'f3.txt']
for filename in filenames:
    count_words(filename)
```

Bài 4. Tệp nén là một loại lưu trữ chứa một hoặc nhiều tệp đã bị giảm kích thước. Nén tập tin trong các hệ điều hành hiện đại thường khá đơn giản. Viết chương trình nén và giải nén tệp bằng Python

```python
import zipfile as zf

def compress(source_path: str, destination_path: str):
    with zf.ZipFile(source_path, 'w', zf.ZIP_DEFLATED) as z:
        z.write(destination_path)
        
def extract(source_path: str, destination_path: str):
    with zf.ZipFile(source_path, 'r', zf.ZIP_DEFLATED) as z:
        z.extract(destination_path)

# compress("source/chap_2/6_image.png", "source/chap_2/6_image.zip")
extract("source/chap_2/6_image.zip", "source/chap_2/6_image.png")
```
