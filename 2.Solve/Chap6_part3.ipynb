```python
import pandas as pd

data1 = pd.Series([5,7,5,1,6], name='seri_example1')
data2 = pd.Series(['C++','Java','Python'], 
                  index=list("CJP"),
                  name='seri_example2')

# Get rows
print(data1[0])
print(data1[1])
print(data1[2])
print(data1[3])
print(data1[4])
print(data1[2:3])       # position slicing (Python base)
print(data1.loc[2:3])
print(data1[data1.values>5])
print(data1[data1.between(3,6)])

print(data2['C':'J'])   # label slicing (same as loc)
print(data2[['C','J']])
print(data2[0:2])
```

```python
# Drop a row
data1 = pd.Series([5,7,5,1,6], name='seri_example1')
print(data1.drop(2))
data1 = pd.Series([5,7,5,1,6], name='seri_example1')
print(data1.drop([2,4]))
```

```python
# Insert and Sort a row
data1 = pd.Series([5,7,5,1,6], name='seri_example1')
data1[2.5]=9
print(data1)
data1.sort_index(inplace=True)
print(data1)
data1.reset_index(drop=True, inplace=True)
print(data1)
```

```python
# Calculations and statistics Method
data1 = pd.Series([5,7,5,1,6], name='seri_example1')
data2 = pd.Series(['C++','Golang', 'Java','Python', 'Swift'], 
                  index=list("CGJPS"),
                  name='Programming Language2')
# print(data1)
print(data1.min())
print(data1.max())
print(data1.mean())
print(data1.idxmax())
print(data1.sum())
print(data1.std())
print(data1.var())
print(data1.argmax())

# print(data2)
print(data2.str.count('Java'))
print(data2.str.count('a'))
```

```python
# String Methods
data2 = pd.Series(['C++','Golang', 'Java','Python', 'Swift'], 
                  index=list("CGJPS"),
                  name='Programming Language2')
print(data2.str.upper())
print(data2.replace('Java','C#'))
print(data2)
```

```python
# Addition between two series
data1 = pd.Series([5,7,5],
                  index=list(['C++','Java','Python']),
                  name='data1')
data2 = pd.Series([15,27,5], 
                  index=list(['C++','Java','Python']),
                  name='data2')
sum = data1 + data2
print(sum)
```

```python
# Addition between two series
data1 = pd.Series([5,7],
                  index=list(['C++','Java']),
                  name='data1')
data2 = pd.Series([15,27,5], 
                  index=list(['C++','Java','Python']),
                  name='data2')
sum = data1 + data2
print(sum)
```

```python
# Group values
data1 = pd.Series(
    [4,1,0,2,2,1], 
    index=list(['C++','Golang','Java','Golang','Java', 'C++']),
    name='data1'
)
print(data1['C++'])
print(data1['Golang'])
print(data1['Java'])
print(data1['C++'].sum())
print(data1['Golang'].sum())
print(data1['Java'].sum())
print(data1.groupby(level=0).sum())
print(data1.groupby(data1>3).sum())
```

```python
# Missing values
data1 = pd.Series([1,6,3,8,None,7,None,2], name='seri_example1')
print(data1)

print(data1.isnull())
print(data1.notnull())
```

```python
# Missing values
print(data1.dropna())
print(data1.fillna(100))
print(data1.interpolate())
```
